__Se cargan las librerias a utilizar__

```python
# se importan las librerias a utilizar.

# Se utiliza pandas para el proceso de limpieza de los datos.
import pandas as pd
import numpy as np
```

__Se cargan los dataframe desde el link del repo__

```python
#se procede a cargar lo datos con la función creada anteriormente y creamos los dataframe.

df_Amazon = pd.read_csv("https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/amazon_prime_titles.csv")
df_Disney = pd.read_csv("https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/disney_plus_titles.csv")
df_Hulu = pd.read_csv("https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/hulu_titles.csv")
df_Netflix= pd.read_json("https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/netflix_titles.json")
```

__Creamos una copia__

```python
#Creamos una copia de los dataframme.
df1_amazon=df_Amazon.copy()
df1_Disney=df_Disney.copy()
df1_Hulu=df_Hulu.copy()
df1_Netflix=df_Netflix.copy()
```

__Borramos los duplicados__

```python
df1_amazon=df1_amazon.drop_duplicates()
df1_Disney=df1_Disney.drop_duplicates()
df1_Hulu=df1_Hulu.drop_duplicates()
df1_Netflix=df1_Netflix.drop_duplicates()
```

__Agregamos una columna que diga el tipo de plataforma__

```python
df1_amazon["Plataforma"]="Amazon"
df1_Disney["Plataforma"]="Disney"
df1_Hulu["Plataforma"]="Hulu"
df1_Netflix["Plataforma"]="Netflix"
```

__Creamos ID compuestos según su plataforma__

Amazon="am"

Disney="di"

Hulu="hu"

Netflix="ne"

```python
df1_amazon["show_id"]=df1_amazon["show_id"]+"am"
df1_Disney["show_id"]=df1_Disney["show_id"]+"di"
df1_Hulu["show_id"]=df1_Hulu["show_id"]+"hu"
df1_Netflix["show_id"]=df1_Netflix["show_id"]+"ne"
```

__Se procede a concatenar los 4 Dataframe para poder trabajarlos en conjunto__

```python
Frames=[df1_amazon,df1_Disney,df1_Hulu,df1_Netflix]
df_Grupal=pd.concat(Frames)
df_Grupal.head(2)
```

```python
# se necesita saber el tipo de dato y la cantidad de nulos que tiene cada columna para poder tomar decisiones.
df_Grupal.info()
```

```python
#creamos un nuevo dataframe que va a tener separados los datos de la columna "duration"
df_dur=df_Grupal["duration"].str.split(expand=True)

#nombramos cada columna.
df_dur.rename(columns={0:"Duracion",1:"Tipo (min-Seasons)"},inplace=True)

#para cambiar el tipo de dato de la columna "duration" a int necesitamos cambiar los campos nulos por ceros(0).
df_dur["Duracion"]=df_dur["Duracion"].fillna(0)
df_dur["Tipo (min-Seasons)"]=df_dur["Tipo (min-Seasons)"].fillna(0)

#Se cambia el tipo de dato de "duracion" a int.
df_dur["Duracion"]=df_dur["Duracion"].astype("int64")

#se comprueba el cambio del tipo de dato
df_dur.info()
```

__Concatenamos los dataframe y eliminamos las columnas viejas__

```python
# se concatena los dos dataframe
df_Grupal=pd.concat([df_Grupal,df_dur],axis=1)

#Se elimina las columnas duration, date_added, rating, country, description, ya que no son relevantes para las consultas
#que vamos a realizar
df_Grupal=df_Grupal.drop(columns=["duration","date_added","rating","country","description"])
```

```python
#completamos las columnas director y cast por "sin dato"
df_Grupal["director"]=df_Grupal["director"].fillna("sin dato")
df_Grupal["cast"]=df_Grupal["cast"].fillna("sin dato")
#usamos show_id como indice.
df_Grupal.set_index(df_Grupal["show_id"],inplace=True)
df_Grupal.head(2)
```

```python
# Cambiar el nombre de las columnas del DF unificado asi podemos identificar rápidamente el contenido.
df_Grupal.rename(columns={"show_id":"id",
                        'type':'Tipo_Film',
                        'title':'Titulo',
                        'director':'Director',
                        'cast':'Reparto',
                        'release_year':'Año_Rodaje',
                        'listed_in':'Genero',
                        'plataform':'Plataforma'},
               inplace=True)
```

__Sacamos las medias de los datos de la columna "Duracion" para movie y TV show, segun su plataforma,
  para posteriormente cambiar los ceros por estas medidas encontradas__

```python
#Sacamos las medias de los datos de la columna "Duracion" para movie y TV show, segun su plataforma,
#para posteriormente cambiar los ceros por estas medidas encontradas
mean_movies = {}
mean_series = {}
for plataform in df_Grupal['Plataforma'].unique():
    mean_movies[plataform] = round(df_Grupal[(df_Grupal['Duracion'] !=0 ) & (df_Grupal['Tipo_Film']=='Movie')&(df_Grupal['Plataforma']==plataform)]['Duracion'].mean())
    mean_series[plataform] = round(df_Grupal[(df_Grupal['Duracion'] !=0 ) & (df_Grupal['Tipo_Film']=='TV Show') & (df_Grupal['Plataforma']==plataform)]['Duracion'].mean())
  
     
#Se hace el cambio de los ceros por las medidass encontradas, la cuales estan calculadas según el tipo de pltaforma
# y segun el tipo de film.
    df_Grupal['Duracion'][df_Grupal['Duracion'].loc[(df_Grupal['Duracion'] == 0) & (df_Grupal['Tipo_Film']=='Movie') & (df_Grupal['Plataforma']==plataform)].index]=mean_movies[plataform]
    df_Grupal['Duracion'][df_Grupal['Duracion'].loc[(df_Grupal['Duracion'] == 0) & (df_Grupal['Tipo_Film']=='TV Show') & (df_Grupal['Plataforma']==plataform)].index]=mean_series[plataform]
print(mean_movies, mean_series)


#Se cambian los valores "sin dato" de las unidades de tiempo (minutos o seasons) en la columna correspondiente
df_Grupal['Tipo (min-Seasons)'][df_Grupal['Tipo (min-Seasons)'].loc[(df_Grupal['Tipo_Film']=='TV Show')].index]='season'
df_Grupal['Tipo (min-Seasons)'][df_Grupal['Tipo (min-Seasons)'].loc[(df_Grupal['Tipo_Film']=='Movie')].index] = 'min'


#Eliminamos la columna de show id de df_grupal.
#df_Grupal=df_Grupal.drop(columns=["show_id"])
```

__Se pasa a minuscula todos los datos para no tener problemas al momento de realizar las consultas__

```python
df_Grupal['Tipo_Film']=df_Grupal['Tipo_Film'].str.lower()
df_Grupal['Titulo']=df_Grupal['Titulo'].str.lower()
df_Grupal['Director']=df_Grupal['Director'].str.lower()
df_Grupal['Reparto']=df_Grupal['Reparto'].str.lower()
df_Grupal['Genero']=df_Grupal['Genero'].str.lower()
```

```python
df_Grupal.head(2)
```

__Se comienza a trabajar con las columna Reparto__

```python
#Convertimos los datos de la columna de Reparto a listas y despues se borra la original.
#Se convierten los actores en listas dentro del dataframe
df_Grupal['Reparto_c'] = df_Grupal["Reparto"].str.split(",")
df_Grupal['Genero_c'] = df_Grupal["Genero"].str.split(",")

#se borra la columna de Reparto y Genero originales
df_Grupal.drop(columns=['Reparto'],inplace=True)
df_Grupal.drop(columns=['Genero'],inplace=True)


#cambiamos nuevamente el nombre a Reparto y Genero para no tener confusion
df_Grupal.rename(columns={'Reparto_c':'Reparto'},inplace=True)
df_Grupal.rename(columns={'Genero_c':'Genero'},inplace=True)
```

__Se crea un Dataframe para los actores__

```python
#Se crea un dataframe con todos los actores
dic_actores={"id":[],"Actor":[],"Año_Rodaje":[],"Plataforma":[]}
for index,reparto in enumerate(df_Grupal['Reparto']):
    for actor in reparto:
        #Se crea un condicional para no cargar los nombres con numero o codigos.
        if len(actor)>3 and actor !="sin dato":
        
        #Se caraga la columna id para poder relacionar
            dic_actores["id"].append(df_Grupal["id"][index])
        #aca se carga los nombres de los actores 
            dic_actores["Actor"].append(actor)
        #Se carga el año de rodaje segun el indice que se esta iterando
            dic_actores["Año_Rodaje"].append(df_Grupal["Año_Rodaje"][index])
        #Se carga la plataforma donde aparece cada actor
            dic_actores["Plataforma"].append(df_Grupal["Plataforma"][index])
    

# se crea el Dataframe de los actores
df_actor = pd.DataFrame(dic_actores)

#comprobamos que el Dataframe cargue correctamente
df_actor.head(2)
```

__Creamos un Dataframe para la columna Genero__

```python
#Creamos un diccionario con los datos que van a conformar el Dataframe.
dic_genero={"id":[],"Genero":[],"Año_Rodaje":[],"Plataforma":[]}
for index,genero in enumerate(df_Grupal['Genero']):
    for gen in genero:
        #Se crea un condicional para no cargar los nombres con numero o codigos.
        if len(gen)>3:
        
        #Se caraga la columna id para poder relacionar
            dic_genero["id"].append(df_Grupal["id"][index])
        #aca se carga los nombres de los actores 
            dic_genero["Genero"].append(gen)
        #Se carga el año de rodaje segun el indice que se esta iterando
            dic_genero["Año_Rodaje"].append(df_Grupal["Año_Rodaje"][index])
        #Se carga la plataforma donde aparece cada actor
            dic_genero["Plataforma"].append(df_Grupal["Plataforma"][index])
    

# se crea el Dataframe de los actores
df_Genero = pd.DataFrame(dic_genero)

#comprobamos que el Dataframe cargue correctamente
df_Genero.head(2)
```


__Exportamos los Dataframe como CSV__
